## **Preprocess**

## **Normalization**

```python
print('Example of Normalization process: ⬇️ \n')
sentence = "عراق سال 2012 قراردادی نظامي در خانه ی مَن به ارزش 4.2 میلیارد دلار با روسیه امضا  کرد."
normalizer = Normalizer()
print(f'Sentence:\n{sentence}\n')
norm_sentence = normalizer.normalize(sentence)
normalizer = Normalizer()
print(f'Sentence after Normalization:\n{norm_sentence}\n')



for i in range(1,len(df[2])):
    df[2][i] = normalizer.normalize(df[2][i])
df
```

## **Remove Stopwords**

```python
#this process needs some time, please wait :)
print('Example of Remove Stopwords process: ⬇️ \n')
print(f'Sentence:\n{norm_sentence}\n')
tokenizer = WordTokenizer()
tokens = tokenizer.tokenize(norm_sentence)
print(f'Tokens:\n{tokens}\n')

# print(tokenizer.tokenize(df[2][1]))
print('All stopwords from the sentence which exists in list of the stopwords -> hazm.stopwords_list:')
for each in tokens:
    if each in utils.stopwords_list():
        print(each)
        tokens.remove(each)

# from nltk.tokenize.treebank import TreebankWordDetokenizer
# TreebankWordDetokenizer().detokenize(tokens)
rem_sentence = normalizer.normalize((" ").join(tokens))
print(f'Sentence after Removing Stopwords:\n{rem_sentence}\n')



for i in range(1,len(df[2])):
    tokens = tokenizer.tokenize(df[2][i])
    tokens = [token for token in tokens if not token in utils.stopwords_list() and  len(token) > 2]
    df[2][i] = normalizer.normalize((" ").join(tokens))
    print(f"content {i} done")
df
```

## **Stemming**

```python
print('Example of Stemming process: ⬇️ \n')

print(f'Sentence:\n{rem_sentence}\n')

tokenizer = WordTokenizer()
tokens = tokenizer.tokenize(rem_sentence)
print(f'Tokens:\n{tokens}\n')

stemmer = Stemmer()

stem_token = []
for each in tokens:
    stem_token.append(stemmer.stem(each))
print(f'Tokens after Stemming:\n{stem_token}\n')

stem_sentence = normalizer.normalize((" ").join(stem_token))
print(f'Sentence after Stemming:\n{stem_sentence}\n')

for i in range(1,len(df[2])):
    tokens = tokenizer.tokenize(df[2][i])
    stem_token = []
    for each in tokens:
        stem_token.append(stemmer.stem(each))
    df[2][i] = normalizer.normalize((" ").join(stem_token))
    # print(f"content {i} done")
df
```

## **Lemmatization**

```python
print('Example of Lemmatization process: ⬇️ \n')

print(f'Sentence:\n{rem_sentence}\n')

tokenizer = WordTokenizer()
tokens = tokenizer.tokenize(rem_sentence)
print(f'Tokens:\n{tokens}\n')

lemmatizer = Lemmatizer()

lemm_token = []
for each in tokens:
    lemm_token.append(lemmatizer.lemmatize(each))
print(f'Tokens after Lemmatization:\n{lemm_token}\n')

lemm_sentence = normalizer.normalize((" ").join(lemm_token))
print(f'Sentence after Lemmatization:\n{lemm_sentence}\n')

for i in range(1,len(df[2])):
    tokens = tokenizer.tokenize(df[2][i])
    lemm_token = []
    for each in tokens:
        lemm_token.append(lemmatizer.lemmatize(each))
    df[2][i] = normalizer.normalize((" ").join(lemm_token))
    print(f"content {i} done")
df
```

## **calculate tf-idf (method one)**

```python
#first of all, we append the Query to dataset and crate new pandas dataframe
Alldata = pd.DataFrame(df[2].append(Query[0]))
Alldata = Alldata.reset_index().drop('index', axis=1)
Alldata[0]
#2428 ---- 2437
```

```python
from sklearn.feature_extraction.text import TfidfTransformer, TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer



words_set = set()

for doc in  Alldata[0]:
    words = tokenizer.tokenize(doc)
    words_set = words_set.union(set(words))
    
print('Number of words in the dataset:',len(words_set))
print('The words in the dataset: \n', words_set)

n_docs = len(Alldata[0])         #·Number of documents in the corpus
n_words_set = len(words_set) #·Number of unique words in the 

df_tf = pd.DataFrame(np.zeros((n_docs, n_words_set)), columns=words_set)

# Compute Term Frequency (TF)
for i in range(n_docs):
    words = tokenizer.tokenize(Alldata[0][i]) # Words in the document
    for w in words:
        df_tf[w][i] = df_tf[w][i] + (1 / len(words))
    # print(f'{i} done')  
df_tf
```

```python

idf = {}

for w in words_set:
    k = 0    # number of documents in the dataset that contain this word
    
    for i in range(n_docs):
        if w in tokenizer.tokenize(Alldata[0][i]):
            k += 1
            
    idf[w] =  np.log10(n_docs / k)
    
df_tf_idf = df_tf.copy()

for w in words_set:
    for i in range(n_docs):
        df_tf_idf[w][i] = df_tf[w][i] * idf[w]
        
df_tf_idf
```

## **calculate tf-idf (method two)**

```python
tr_idf_model  = TfidfVectorizer()
tf_idf_vector = tr_idf_model.fit_transform(Alldata[0])
print(type(tf_idf_vector), tf_idf_vector.shape)
tf_idf_array = tf_idf_vector.toarray()
print(tf_idf_array)
words_set = tr_idf_model.get_feature_names()
print(words_set)
df_tf_idf = pd.DataFrame(tf_idf_array, columns = words_set)
df_tf_idf
```

```python
from numpy.ma.core import append
import numpy as np
from numpy.linalg import norm
allcosine = []

for query in range(2428,2438):
    cos = []
    for i in range(len(df[2])):
        A = np.array(df_tf_idf[i:i+1])[0]
        B = np.array(df_tf_idf[query:query+1])[0]
        cosine = np.dot(A,B)/(norm(A)*norm(B))
        cos.append(cosine)
    allcosine.append(cos)
allcosine
```

## **The top 10  docs with highest similarity cosine with query**

```python

sim = pd.DataFrame (allcosine).transpose()
sim.columns = ['Q1','Q2','Q3','Q4','Q5','Q6','Q7','Q8','Q9','Q10']
sim
```
